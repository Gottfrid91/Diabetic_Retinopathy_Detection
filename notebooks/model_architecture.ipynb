```python
import tensorflow as tf
slim = tf.contrib.slim
```

# Layers loaded model

```python
# Let's load a previously saved meta graph in the default graph
# This function returns a Saver
path = '/home/olle/PycharmProjects/Diabetic_Retinopathy_Detection/output/pre_weights'
saver = tf.train.import_meta_graph(path + '/model.ckpt-1030928.meta')
```

```python
# We can now access the default graph where all our metadata has been loaded
graph = tf.get_default_graph()
sess=tf.Session()    

saver.restore(sess,tf.train.latest_checkpoint(path))
variables_to_restore = slim.get_variables_to_restore()
variables_to_restore
```
